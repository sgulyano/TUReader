## รามข้อมูล

เปิดไฟล์ทั้งสองส่วนที่ถูกแก้ไขเบื้องต้นแล้ว

```python
df1 = pd.read_csv('TUlist.csv', index_col=0)
df2 = pd.read_csv('MUAlist.csv', index_col=0)
df3 = pd.read_csv('Councillist.csv', index_col=0)
```

## แก้รายชื่อซ้ำ
หารายชื่อทั้งหมดที่ไม่ซ้ำกัน

```python
u, uid, uin, uc = np.unique(df[['ชื่อ', 'นามสกุล']].fillna('').apply(lambda x: ' '.join(x), axis=1), 
                            return_index=True, 
                            return_inverse=True, 
                            return_counts=True)
```

หา boolean index (list ของ True/False ความยาวเท่าจำนวนข้อมูลรายชื่อ) ของรายชื่อที่จะเก็บไว้ โดยทำดังนี้
* หาชื่อที่ไม่มีซ้ำ แล้วไปค้นหาตำแหน่งในรายชื่อจริง
* ส่วนชื่อที่มีซ้ำ ให้เก็บบรรทัดล่างสุด (มันจะเป็นข้อมูลใหม่สุด รู้ได้จากตำแหน่งวิชาการที่สูงกว่า)

```python
idx1 = np.where(uc == 1)[0]

arr = []
for i in np.where(uc > 1)[0]:
    arr.append(np.where(uin == i)[0].max())
    
id1 = (uin[:,np.newaxis]==idx1).any(axis=1)
id1[arr] = True
```

เลือกรายชื่อที่ไม่มีข้อมูลซ้ำ

```python
df2 = df[id1]
```

#### Sanity Check
ว่าการแก้รายชื่อซ้ำสำเร็จ

```python
u2, _, _, uc2 = np.unique(df2[['ชื่อ', 'นามสกุล']].fillna('').apply(lambda x: ' '.join(x), axis=1), 
                            return_index=True, 
                            return_inverse=True, 
                            return_counts=True)
np.where(uc2 > 1)[0]
```
